```python
features = ['Source Port', 'Destination Port', 'Packet Length', 'Anomaly Scores']
X = df[features]
X
```

```python
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
inertias = []
k_range = range(2, 12)
for i in k_range:
    kmeans = KMeans(n_clusters=i, random_state=42, init='k-means++', n_init=10)
    kmeans.fit(X_pca)
    inertias.append(kmeans.inertia_)
plt.plot(k_range, inertias)
```

```python
from kneed import KneeLocator
knee_point = KneeLocator(k_range, inertias, curve='convex', direction='decreasing')
knee_point.knee
```

```python
kmeans = KMeans(n_clusters=4, random_state=0) 
kmeans.fit(X_pca)
```

```python
plt.scatter(X_pca[:, 0],X_pca[:, 1], c=kmeans.labels_, cmap='viridis')
plt.title("K-Means Clustering")
plt.show()
```
