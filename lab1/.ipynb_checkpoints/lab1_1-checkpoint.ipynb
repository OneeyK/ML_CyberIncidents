```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
from sklearn.model_selection import GridSearchCV
import matplotlib.pyplot as plt
```

```python
file_path = 'uci_malware_detection.csv'  
df = pd.read_csv(file_path)
```

```python
X = df.drop(["Label"],axis=1)
y = LabelEncoder().fit_transform(df['Label'])
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
param_grid = {'C': [0.01, 0.1, 1, 10, 100],'penalty': ['l1', 'l2'],'solver': ['liblinear']}
grid_search = GridSearchCV(LogisticRegression(), param_grid, cv=5, scoring='accuracy')
grid_search.fit(X_train, y_train)
```

```python
print("Найкращі параметри регуляризації:", grid_search.best_params_)
```

```python
grid_search.fit(X_train, y_train)
```

```python
best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)
```

```python
print("Classification Report:\n", classification_report(y_test, y_pred))
print("Accuracy:\n", accuracy_score(y_test, y_pred))
```

```python
results = grid_search.cv_results_

# Plot accuracy for different values of C
plt.plot(results['param_C'], results['mean_test_score'])
plt.xscale('log')
plt.xlabel('C (Log Scale)')
plt.ylabel('Mean Accuracy')
plt.title('Effect of Regularization Strength on Accuracy')
plt.show()
```
